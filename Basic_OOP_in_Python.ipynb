#Task 1: Create a class and object


```python
class Product:
  def __init__(self, name, price):
    self.name = name
    self.price = price

item = Product("Laptop", 1000)
print(item.name, item.price)
```

#Task 2:Add method to class

```python
class Product:
  def __init__(self, name, price):
    self.name = name
    self.price = price

  def discount(self, percent):
    return self.price * (1 - percent/100)

item = Product("Laptop", 1000)
print(item.discount(10)) #900
```

#Task 3: Multiple objects

```python
item1 = Product("Laptop", 1000)
item2 = Product("Phone", 700)

print(item1.discount(10)) #900
print(item2.discount(20)) #500
```

#Task 4: Real Data Science Use case-Feature object

```python
class Feature:
  def __init__(self, values):
    self.values = values

  def normalize(self):
    min_val =min(self.values)
    max_val =max(self.values)
    return [(x - min_val) / (max_val - min_val) for x in self.values]

data_feature = Feature([10, 20, 30, 40])
print(data_feature.normalize())
```

#Task 5:Handing invallid data with oop

```python
class Feature:
  def __init__(self, values):
    self.values = values
  def clean_ints(self):
    clean_data = []
    for x in self.values:
      try:
        clean_data.append(int(x))
      except ValueError:
        continue
    return clean_data



feature = Feature(["10", "20","abs", "30"])
print(feature.clean_ints())
```

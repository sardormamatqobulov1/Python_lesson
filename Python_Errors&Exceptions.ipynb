#Test 6: Handle exceptions

```python
try:
  num = int(input("Enter a number: "))
except ValueError:
  print("Invalid input, please enter a number.")
```

#Task 7:Real Data science use case

```python
data = ["10", "20", "abc", "30"]
clean_data = []

for x in data:
  try:
    clean_data.append(int(x))
  except ValueError:
    continue

print(clean_data)
```

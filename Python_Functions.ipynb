#Task 2: Calculate mean(average)

```python
def mean(data):
  total = 0
  for x in data:
    total += x
  return total / len(data)
```

```python
sales = [120, 150, 170, 200]
mean(sales)
```

#Task 3:Normalize data

```python
def normalize(data):
  min_val = min(data)
  max_val = max(data)

  normalized = []
  for x in data:
    normalized.append((x - min_val)/(max_val - min_val))
  return normalized
```

```python
score = [50, 60, 70, 80, 90]
normalize(score)
```

#Task 4 : Filter data

```python
def filter_above(data, threshold):
  result = []
  for x in data:
    if x > threshold:
      result.append(x)
  return result
```

```python
temps =[22,25,30, 18, 35]
filter_above(temps, 20)
```

#Task 5 : Featrue engineering

```python
def square_feature(data):
  features = []
  for x in data:
    features.append(x ** 2)
  return features
```

```python
values = [ 1, 2, 3, 4]
square_feature(values)
```

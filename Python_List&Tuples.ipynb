#Task 3: Loop through a list


```python
temperatures = [22, 25, 30, 18, 35]

for t in temperatures:
  print(t)
```

#Task 4: Simple data filtering

```python
values = [5, 12, 8, 20, 3]
filtered = []

for x in values:
  if x > 6:
    filtered.append(x)

print(filtered)
```
